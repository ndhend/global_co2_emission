```python
from pandas._libs.hashtable import value_count
# Graph 2: worst and best changers

import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/ndhend/csv_data_winc/main/co2_emissions.csv'
df = pd.read_csv(url)

# name df
countries = df
# rename columns
countries.columns = ['country','1990', '2005', '2017', '2021']


# Take out the non-countries
countries = countries.loc[3:, ['country','1990', '2005', '2017']]
countries = countries[countries.index != 67]
countries
# relative change:
# Mt 2005 = 2005/1990*100
# Mt 2017 = 2017/1990 *100

# Write functions, creating lists

def calculate_change_05(countries):
  change_2005_list = []
  for index, row in countries.iterrows():
    change_2005 = (row['2005'] / row['1990']) * 100
    change_2005_list.append(change_2005)
  return change_2005_list

# Call the function and assign the result to a new column in the DataFrame
countries['Change_2005'] = calculate_change_05(countries)

def calculate_change_17(countries):
  change_2017_list = []
  for index, row in countries.iterrows():
    change_2017 = (row['2017'] / row['1990']) * 100
    change_2017_list.append(change_2017)
  return change_2017_list

# Call the function and assign the result to a new column in the DataFrame
# new columns relative change

countries['Change_2005'] = calculate_change_05(countries)
countries['Change_2017'] = calculate_change_17(countries)

# sort by 'Change_2017'
sorted_countries_17 = countries.sort_values('Change_2017')

# create the same chart for all countries that
# had at least five Mt of CO2 emissions in 1990.
# Boolean mask countries with >= 5 in 1990
# Met >= 5 kreeg ik nog steeds meer dan 100 rijen
# Ik begrijp de opdracht misshien niet maar ik heb er 500 van gemaakt.
sorted_countries_17['1990']  >= 5
sorted_countries_17 = sorted_countries_17.loc[sorted_countries_17['1990']  >= 5]
sorted_countries_17 = sorted_countries_17.reset_index(drop=True)
# sorted_countries_17


# X-axis: the years 1990, 2005, 2017
# Drop the 'City' column in-place
rel_co2_emission = sorted_countries_17.drop(['2005', '2017'], axis=1)
rel_co2_emission.columns = ['country', '1990','2005', '2017']
rel_co2_emission

fig, ax = plt.subplots()   #create a figure with one or more axis

# first seperate the years for the callable (see smoking in london 2)
years = rel_co2_emission.columns[1:]

for index, row in rel_co2_emission.iterrows():
  plt.plot(years, row[1:], label=row[0])

# in the callable we stated that the labels are the countries (row[0])

plt.xlabel("years") # label for x-axis
plt.ylabel("relative change in co2 emission") # label for y-axis
plt.title("Countries with relative change in co2 emission") # set title of plot

# place the legend box outside the graph
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))


plt.show()
# y-axis: relative amount 1990, 2005, 2017
# get list of countries with relative change
```
